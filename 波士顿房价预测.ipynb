```python
x = fangzi.drop(columns = '房价中位数')
x
```

```python
from sklearn.feature_selection import SelectKBest     #选择排名前k个的特征
from sklearn.feature_selection import f_regression   
#选择的统计指标须要指定，对于regression问题，使用f_regression指标;对于classification问题，可使用chi2或者f_classif指标
y=fangzi['房价中位数']
SelectKBest = SelectKBest(f_regression,k=3)
bastFeature=SelectKBest.fit_transform(x,y)
SelectKBest.get_support()
x.columns[SelectKBest.get_support()]
```

```python
#特征归一化，此处是最大值最小值归一化:每个数据减去最小值再除以这列的最大值
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
for feature in fangzi.columns:
    fangzi['标准化'+feature]=scaler.fit_transform(fangzi[[feature]])
fangzi.head()
```

```python
from sklearn.model_selection import train_test_split#划分测试集和训练集
from sklearn.linear_model import LinearRegression#线性回归
from sklearn import metrics#指标
import warnings#忽略一些警告
warnings.filterwarnings('ignore')
```

```python
X=x
X.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred_class=model.predict(X_test)
y_pred_class
```

```python
#平均绝对误差
metrics.mean_absolute_error(y_test,y_pred_class)
```

```python
#均方误差：（每个值－估计值）的平方和的平均值
metrics.mean_squared_error(y_test,y_pred_class)
```

```python
#均方误差 竞赛常用
from math import sqrt
print(sqrt(metrics.mean_squared_error(y_test,y_pred_class)))
```

```python
#均方根对数误差
#对欠预测的惩罚更大，例如共享单车，欠预测意味着单车数量将不够
metrics.mean_squared_log_error(y_test,y_pred_class)
```

```python
#R平方
metrics.r2_score(y_test,y_pred_class)
```

```python
### 导出
import pandas as pd
result = {'prediction':model.predict(X_test)}
result
```

```python
result_file = pd.DataFrame(result)
result_file
```

```python
result_file.to_csv('linear_regression.csv')
```

### 随机森林处理数据

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators':[5,10,20,50,100,200],
    'max_depth':[3,5,7],
    'max_features':[0.6,0.7,0.8,1]
}

rf = RandomForestRegressor()
grid = GridSearchCV(rf,param_grid = param_grid,cv = 3)

grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# 制定模型为效果最好参数对应的模型
rf_reg = grid.best_estimator_
rf_reg
```

## 可视化其中一棵决策树

```python
from sklearn import tree 
import pydotplus
from IPython.display import Image,display
#从随机森林模型中选取一颗决策树进行可视化
estimator=rf_reg.estimators_[3]
dot_data=tree.export_graphviz(estimator,
                              out_file=None,
                              #feature names=iris feature name，
                              #class names =iris target name，
                              filled=True,
                              rounded=True
                             )
graph =pydotplus.graph_from_dot_data(dot_data)
display(Image(graph.create_png()))
```

```python
import numpy as np

print('特征排序：')
feature_name = X.columns
feature_importance = rf_reg.feature_importances_
indices = np.argsort(feature_importance)[::-1]
indices 
feature_importance[indices]
# for index in indices:
#     print('feature {} 的重要性为{}'.format(feature_name[index],feature_importance[index]))
```

```python
plt.figure(figsize = (20,8))
plt.title('随机森林模型不同特征的重要度')
plt.bar(range(len(feature_importance)),feature_importance[indices],color ='orange')
plt.xticks(ticks = range(len(feature_importance)),labels = feature_name[indices],color = 'red',rotation=45)
plt.show()
#ticks表示刻度，labels表示刻度上的标签
```
