```python
from sklearn.datasets import make_circles,make_moons,make_blobs,make_classification
```

```python
n_samples = 100

datasets = [make_circles(n_samples = n_samples,noise = 0.2,factor = 0.5,random_state = 1),
           make_moons(n_samples = n_samples,noise = 0.2,random_state = 0),
           make_blobs(n_samples = n_samples,centers = 2,random_state = 5),
           make_classification(n_samples = n_samples,n_features = 2,n_informative = 2,random_state = 5
                              ,n_redundant = 0)]
```

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from  matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
```

### ,45

```python
kernel = ['linear','poly','rbf','sigmoid']


row = len(kernel)
col = len(kernel) + 1

fig,ax = plt.subplots(row,col,figsize = (20,16))
```

### 4

```python
for indx,(X,y) in enumerate(datasets):
    #
    axes = ax[indx,0]
    if indx == 0:
        axes.set_title('Input data')      #
    axes.scatter(X[:,0],X[:,1],c = y,zorder = 10,cmap = plt.cm.Paired,edgecolor = 'k')
    axes.set_xticks([])
    axes.set_yticks([])
```

```python
fig
```

### 

```python

for n_index,n_kernel in enumerate(kernel):
    clf = SVC(kernel = n_kernel,gamma = 2)
    clf = clf.fit(X,y)     #fit
    score = clf.score(X,y)
    
    axes1 = ax[0,n_index + 1]
    
    #
    axes1.scatter(X[:,0],X[:,1],c = y,zorder = 10,cmap = plt.cm.Paired,edgecolor = 'k')
    
    #
    axes1.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s = 50
                 ,facecolors = 'none',zorder = 10,edgecolor = 'k')    #facecolors = 'none'
    
    #
    X_min,X_max = X[:,0].min() - .5 , X[:,0].max() + .5
    y_min,y_max = X[:,1].min() - .5 , X[:,1].max() + .5
    
    #np.mgrid,np.linspacenp.meshgrid
    #
    #[]
    #
    XX,yy = np.mgrid[X_min:X_max:200j,y_min:y_max:200j]
    #np.c_np.vstack
    Z = clf.decision_function(np.c_[XX.ravel(),yy.ravel()]).reshape(XX.shape)
    #
    axes1.pcolormesh(XX,yy,Z > 0,cmap = plt.cm.Paired,shading='auto')
    #
    axes1.contour(XX,yy,Z,colors = ['k','k','k'],linestyles = ['--','-','--'],
              levels = [-1,0,1])

    axes1.set_xticks([])
    axes1.set_yticks([])   #

        
    #
    if n_index == 0:
        axes1.set_title(n_kernel)
    
    #
    axes1.text(0.95,0.16,('%.2f' % score).lstrip('0')
              ,size = 15
              ,bbox = dict(boxstyle = 'round',alpha = 0.8,facecolor = 'white')
              #
              ,transform = axes1.transAxes  #ax
              ,horizontalalignment = 'right'      #
              )
    
```

```python
fig,ax = plt.subplots(row,col,figsize = (20,16))

for indx,(X,y) in enumerate(datasets):
    #
    ax1 = ax[indx,0]
    if indx == 0:
        ax1.set_title('Input data')      #
    ax1.scatter(X[:,0],X[:,1],c = y,zorder = 10,cmap = plt.cm.Paired,edgecolor = 'k')
    
    ax1.set_xticks([])
    ax1.set_yticks([])

    
    for n_index,n_kernel in enumerate(kernel):
        clf = SVC(kernel = n_kernel,gamma = 2)
        clf = clf.fit(X,y)     #fit
        score = clf.score(X,y)

        axes2 = ax[indx,n_index + 1]

        #
        axes2.scatter(X[:,0],X[:,1],c = y,zorder = 10,cmap = plt.cm.Paired,edgecolor = 'k')

        #
        axes2.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s = 50
                     ,facecolors = 'none',zorder = 10,edgecolor = 'k')    #facecolors = 'none'

        #
        X_min,X_max = X[:,0].min() - .5 , X[:,0].max() + .5
        y_min,y_max = X[:,1].min() - .5 , X[:,1].max() + .5

        #np.mgrid,np.linspacenp.meshgrid
        #
        #[]
        #
        XX,yy = np.mgrid[X_min:X_max:200j,y_min:y_max:200j]
        #np.c_np.vstack
        Z = clf.decision_function(np.c_[XX.ravel(),yy.ravel()]).reshape(XX.shape)
        #
        axes2.pcolormesh(XX,yy,Z > 0,cmap = plt.cm.Paired,shading='auto')
        #
        axes2.contour(XX,yy,Z,colors = ['k','k','k'],linestyles = ['--','-','--'],
                  levels = [-1,0,1])

        axes2.set_xticks([])
        axes2.set_yticks([])   #


        #
        if indx == 0:
            axes2.set_title(n_kernel)

        #
        axes2.text(0.95,0.16,('%.2f' % score).lstrip('0')
                  ,size = 15
                  ,bbox = dict(boxstyle = 'round',alpha = 0.8,facecolor = 'white')
                  #
                  ,transform = axes2.transAxes  #ax
                  ,horizontalalignment = 'right'      #
                  )

    
plt.tight_layout()    #
plt.show()
```
