### 对四种不同的数据集，采用4种不同的核函数对数据集进行划分

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from  matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
```

### 建立子图，创建画布,4行5列

```python
kernel = ['linear','poly','rbf','sigmoid']


row = len(kernel)
col = len(kernel) + 1

fig,ax = plt.subplots(row,col,figsize = (20,16))
```

### 在每行的第一列展示4种数据集的数据分布

```python
for indx,(X,y) in enumerate(datasets):
    #在图像的第一列，放置原数据的分布
    axes = ax[indx,0]
    if indx == 0:
        axes.set_title('Input data')      #在最上面对该列图像进行标签处理
    axes.scatter(X[:,0],X[:,1],c = y,zorder = 10,cmap = plt.cm.Paired,edgecolor = 'k')
    axes.set_xticks([])
    axes.set_yticks([])
```

```python
fig
```

### 在第一行对不同的核函数处理后的数据进行绘图

```python

for n_index,n_kernel in enumerate(kernel):
    clf = SVC(kernel = n_kernel,gamma = 2)
    clf = clf.fit(X,y)     #fit的结果就是寻找出决策边界
    score = clf.score(X,y)
    
    axes1 = ax[0,n_index + 1]
    
    #绘制图像本身的散点图
    axes1.scatter(X[:,0],X[:,1],c = y,zorder = 10,cmap = plt.cm.Paired,edgecolor = 'k')
    
    #绘制支持向量
    axes1.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s = 50
                 ,facecolors = 'none',zorder = 10,edgecolor = 'k')    #facecolors = 'none'将该点画成透明的
    
    #绘制决策边界
    X_min,X_max = X[:,0].min() - .5 , X[:,0].max() + .5
    y_min,y_max = X[:,1].min() - .5 , X[:,1].max() + .5
    
    #np.mgrid,合并np.linspace和np.meshgrid两种用法
    #一次性使用最大值和最小值生成网络
    #表示为[起始值：结束值：步长]
    #如果步长是复数，则其整数部分就是起始值和结束值之间创建的点的数量，并且结束值被包含在内
    XX,yy = np.mgrid[X_min:X_max:200j,y_min:y_max:200j]
    #np.c_，类似于np.vstack的功能
    Z = clf.decision_function(np.c_[XX.ravel(),yy.ravel()]).reshape(XX.shape)
    #填充等高线不同区域的颜色
    axes1.pcolormesh(XX,yy,Z > 0,cmap = plt.cm.Paired,shading='auto')
    #绘制等高线
    axes1.contour(XX,yy,Z,colors = ['k','k','k'],linestyles = ['--','-','--'],
              levels = [-1,0,1])

    axes1.set_xticks([])
    axes1.set_yticks([])   #设置坐标轴不显示标签

        
    #将标题放到第一行的顶上
    if n_index == 0:
        axes1.set_title(n_kernel)
    
    #为每张图添加分类的分数
    axes1.text(0.95,0.16,('%.2f' % score).lstrip('0')
              ,size = 15
              ,bbox = dict(boxstyle = 'round',alpha = 0.8,facecolor = 'white')
              #为分数添加一个白色的格子作为底色
              ,transform = axes1.transAxes  #确定文字所对的坐标轴，就是ax子图的坐标轴本身
              ,horizontalalignment = 'right'      #位于坐标轴的什么方向
              )
    
```

### 将所有的数据放到循环中，实现4列5行的图像绘制

```python
fig,ax = plt.subplots(row,col,figsize = (20,16))

for indx,(X,y) in enumerate(datasets):
    #在图像的第一列，放置原数据的分布
    ax1 = ax[indx,0]
    if indx == 0:
        ax1.set_title('Input data')      #在最上面对该列图像进行标签处理
    ax1.scatter(X[:,0],X[:,1],c = y,zorder = 10,cmap = plt.cm.Paired,edgecolor = 'k')
    
    ax1.set_xticks([])
    ax1.set_yticks([])

    
    for n_index,n_kernel in enumerate(kernel):
        clf = SVC(kernel = n_kernel,gamma = 2)
        clf = clf.fit(X,y)     #fit的结果就是寻找出决策边界
        score = clf.score(X,y)

        axes2 = ax[indx,n_index + 1]

        #绘制图像本身的散点图
        axes2.scatter(X[:,0],X[:,1],c = y,zorder = 10,cmap = plt.cm.Paired,edgecolor = 'k')

        #绘制支持向量
        axes2.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s = 50
                     ,facecolors = 'none',zorder = 10,edgecolor = 'k')    #facecolors = 'none'将该点画成透明的

        #绘制决策边界
        X_min,X_max = X[:,0].min() - .5 , X[:,0].max() + .5
        y_min,y_max = X[:,1].min() - .5 , X[:,1].max() + .5

        #np.mgrid,合并np.linspace和np.meshgrid两种用法
        #一次性使用最大值和最小值生成网络
        #表示为[起始值：结束值：步长]
        #如果步长是复数，则其整数部分就是起始值和结束值之间创建的点的数量，并且结束值被包含在内
        XX,yy = np.mgrid[X_min:X_max:200j,y_min:y_max:200j]
        #np.c_，类似于np.vstack的功能
        Z = clf.decision_function(np.c_[XX.ravel(),yy.ravel()]).reshape(XX.shape)
        #填充等高线不同区域的颜色
        axes2.pcolormesh(XX,yy,Z > 0,cmap = plt.cm.Paired,shading='auto')
        #绘制等高线
        axes2.contour(XX,yy,Z,colors = ['k','k','k'],linestyles = ['--','-','--'],
                  levels = [-1,0,1])

        axes2.set_xticks([])
        axes2.set_yticks([])   #设置坐标轴不显示标签


        #将标题放到第一行的顶上
        if indx == 0:
            axes2.set_title(n_kernel)

        #为每张图添加分类的分数
        axes2.text(0.95,0.16,('%.2f' % score).lstrip('0')
                  ,size = 15
                  ,bbox = dict(boxstyle = 'round',alpha = 0.8,facecolor = 'white')
                  #为分数添加一个白色的格子作为底色
                  ,transform = axes2.transAxes  #确定文字所对的坐标轴，就是ax子图的坐标轴本身
                  ,horizontalalignment = 'right'      #位于坐标轴的什么方向
                  )

    
plt.tight_layout()    #将表格紧缩
plt.show()
```

### rbf，高斯径向基核函数基本在任何数据集上都表现不错，属于比较万能的核函数。可以再实际的应用中，先尝试rbf。该核函数适用于核转换到很高空间的情况，在各种情况下效果都很不错。多项式核函数多被用于图像处理之中。
